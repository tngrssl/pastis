```python
p = reco.pipeline("sc")

p.dataset[0]["legend"] = "sLASER 20:1"
p.dataset[0]["resp_bpm"] = 18
p.dataset[0]["heart_bpm"] = 58
p.dataset[0]["dcm"]["files"] = ["/home/tangir/crmbm/acq/319-fc-p1-moelle/20191017/01_0015_slaser-r-n",
                                "/home/tangir/crmbm/acq/319-fc-p1-moelle/20191017/01_0013_slaser-r-n"]
p.dataset[0]["raw"]["files"] = ["/home/tangir/crmbm/acq_twix/319-fc-p1-moelle/meas_MID142_slaser_R_N=20+_1_longTE_SNR++++_FID43720.dat",
                                "/home/tangir/crmbm/acq_twix/319-fc-p1-moelle/meas_MID138_slaser_R_N=20+_1_longTE_SNR++++_FID43716.dat"]

p.settings["display"] = True
p.run()
```

```python
fig = plt.figure(162)
fig.set_size_inches(10, 6)
fig.suptitle("")
fig.axes[0].set_title("Amplitude change")
fig.axes[0].set_ylabel("Amplitude relative change (%)")
fig.axes[1].set_title("FWHM linewidth change")
fig.axes[1].set_ylabel("FWHM linewidth (Hz)")
fig.axes[2].set_title("Chemical shift change")
fig.axes[2].set_ylabel("Chemical shift relative change (ppm)")
fig.axes[2].set_xlabel("Acquisition time (s)")
fig.axes[3].set_title("Phase change")
fig.axes[3].set_ylabel("Phase relative change (rd)")
fig.axes[3].set_xlabel("Acquisition time (s)")

fig.axes[4].set_ylabel("Individual acquired spectrum index")
fig.axes[4].set_xlabel("Chemical shift (ppm)")
```

```python
tmp_fig = plt.figure(132000)
tmpax = tmp_fig.subplots()
lll = tmpax.plot([0, 1], [0, 1], 'green', [0, 1], [0, 1], 'red', [0, 1])

fig.axes[4].legend(lll, ["Accepted", "Rejected"], bbox_to_anchor=(0, 1, 1., .102), loc=3, ncol=2)
fig.tight_layout()

# annotations...
props = dict(boxstyle='square', facecolor='white', alpha=0.9, edgecolor='grey')

xlim = fig.axes[0].get_xlim()
ylim = fig.axes[0].get_ylim()
fig.axes[0].text(xlim[0] - (xlim[0]-xlim[1])*0.04, 
                     ylim[0] + (ylim[1]-ylim[0])*0.06, 
                     "A.",
                     fontsize=12,
                     bbox=props)

xlim = fig.axes[1].get_xlim()
ylim = fig.axes[1].get_ylim()
fig.axes[1].text(xlim[0] - (xlim[0]-xlim[1])*0.04, 
                     ylim[0] + (ylim[1]-ylim[0])*0.06, 
                     "B.",
                     fontsize=12,
                     bbox=props)

xlim = fig.axes[2].get_xlim()
ylim = fig.axes[2].get_ylim()
fig.axes[2].text(xlim[0] - (xlim[0]-xlim[1])*0.04, 
                     ylim[0] + (ylim[1]-ylim[0])*0.06, 
                     "C.",
                     fontsize=12,
                     bbox=props)

xlim = fig.axes[3].get_xlim()
ylim = fig.axes[3].get_ylim()
fig.axes[3].text(xlim[0] - (xlim[0]-xlim[1])*0.04, 
                     ylim[0] + (ylim[1]-ylim[0])*0.06, 
                     "D.",
                     fontsize=12,
                     bbox=props)

xlim = fig.axes[4].get_xlim()
ylim = fig.axes[4].get_ylim()
fig.axes[4].text(xlim[0] - (xlim[0]-xlim[1])*0.04, 
                     ylim[0] + (ylim[1]-ylim[0])*0.04, 
                     "E.",
                     fontsize=12,
                     bbox=props)

fig.savefig("./notebooks/figs/fig_data_rej.svg")
print("S11")
```

# Data rejection of motion data

TWIX data was broken... but ok using matlab's mapvbvd 

```python
import scipy.io
plt.close("all")

s = reco.MRSData2("/home/tangir/crmbm/acq_twix/349-ap-p2-moelle/meas_MID102_slaser_R_N=10_2_longTE_SNR+++_FID63232.dat")
mat = scipy.io.loadmat("/home/tangir/crmbm/acq_twix/349-ap-p2-moelle/fids.mat")
data_from_mat = np.squeeze(mat["ttt"][:, :, 0, 0, 0, 0, 0, 0, 0, :])
data_from_mat = np.transpose(data_from_mat, (2, 1, 0))
s[:] = data_from_mat[:, :, 0:8192]

s_ref = reco.MRSData2("/home/tangir/crmbm/acq_twix/349-ap-p2-moelle/meas_MID101_slaser_R_N=10_2_longTE_SNR+++_FID63231.dat")

s._data_ref = s_ref
s._display_label = ""

#s = s.correct_phase_3d()
s = s.correct_intensity_scaling_nd()
s = s.correct_combine_channels_3d(False)
s = s.correct_zerofill_nd()
s = s.correct_apodization_nd(10)
s = s.correct_realign_2d([4.5, 5], 3)
s = s.correct_analyze_and_reject_2d([4.5, 5], [4.5, 5], [4.5, 5], 9, False,
                                    {"amplitude (%)": None, 
                                     "linewidth (Hz)": None, 
                                     "chemical shift (ppm)": None, 
                                     "phase std. factor (%)": None}, True, [reco.data_rejection_method.AUTO_AMPLITUDE,
                                                reco.data_rejection_method.AUTO_LINEWIDTH,
                                                reco.data_rejection_method.AUTO_FREQUENCY,
                                                reco.data_rejection_method.AUTO_PHASE], 1, 0.0,
                                    True, [1, 6])
print("S17")
```
